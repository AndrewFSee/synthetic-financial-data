```python
import sys
sys.path.insert(0, '../src')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from synfin.data.download import load_ohlcv
from synfin.data.preprocess import compute_log_returns
from synfin.evaluation.stylized_facts import check_all_stylized_facts

sns.set_theme(style='whitegrid')
print('Libraries loaded successfully!')
```

```python
# Load data (run scripts/download_data.py first)
df = load_ohlcv('AAPL', data_dir='../data/raw')
if df is not None:
    df = compute_log_returns(df)
    print(df.describe())
    df[['Close', 'Volume']].plot(subplots=True, figsize=(14, 6), title='AAPL OHLCV')
    plt.tight_layout()
    plt.show()
else:
    print('No data found. Run: python scripts/download_data.py --tickers AAPL')
```
