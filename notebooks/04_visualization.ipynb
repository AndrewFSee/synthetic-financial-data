# 04 â€” Visualization

Candlestick charts, distribution plots, correlation heatmaps, and time series comparisons.

```python
import sys
sys.path.insert(0, '../src')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from synfin.visualization.distributions import plot_distributions
from synfin.visualization.correlations import plot_correlation_heatmap
from synfin.visualization.time_series import plot_time_series

print('Visualization tools loaded!')
```

```python
# Create dummy data
np.random.seed(42)
feature_names = ['Open', 'High', 'Low', 'Close', 'Volume', 'LogReturn', 'LogVolume', 'DollarVolume']
real = np.random.randn(200, 30, 8).astype(np.float32)
synthetic = (np.random.randn(200, 30, 8) * 1.1).astype(np.float32)

# Distribution comparison
print('Plotting distributions...')
plot_distributions(real, synthetic, feature_names=feature_names, show=True)
```

```python
# Correlation heatmaps
print('Plotting correlation heatmaps...')
plot_correlation_heatmap(real, synthetic, feature_names=feature_names, show=True)
```

```python
# Time series comparison
print('Plotting time series...')
plot_time_series(real, synthetic, feature_names=feature_names, num_samples=3, show=True)
```
